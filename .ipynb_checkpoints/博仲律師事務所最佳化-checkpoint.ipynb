```python
# read solar wheeling data
lawer_solar = pd.read_excel("lawer_solar.xlsx", parse_dates=["時間"], index_col='時間')
lawer_solar = lawer_solar[['月服務使用量(kWh)']]
lawer_solar.head()
```

```python
def visualize_group_save(data_mean, col, x_label, y_label, title, ymin=None, ymax=None, name=False, valuemark=False, kind='line'):
    ax = data_mean[col].plot(kind=kind, figsize=(16,8), fontsize=12, linewidth=3, linestyle='-')
    ax.get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
    ax.set_xlabel(x_label, fontsize=12)
    ax.set_ylabel(y_label, fontsize=12)
    ax.set_title(title,fontsize=16)
    ax.set_ylim(ymin=ymin, ymax=ymax)
    if valuemark:
        for index,data in enumerate(data_mean[col]):
            plt.text(x=index , y =data+1 , s=f"{int(data):,}" , fontdict=dict(fontsize=20))
    if isinstance(name, str):
        plt.savefig(name)
    plt.show()
```

```python
lawer_solar['month'] = lawer_solar.index.month
solar_group_month = lawer_solar.groupby('month').sum()
# show solar wheeling
visualize_group_save(solar_group_month, "月服務使用量(kWh)", "月份", "kWh", "博仲律師事務所每月太陽能服務使用量")
```

或許用dummy variable 可以解?

新增變數 $y_{ni}, y_{wi}$ 

台電五項轉供式:
        
$$ G_{mi} = R_{m} \times P_{mi} $$
    
$$ U_{ni} = min [A_{ni} , C_{ni}] $$
    
$$ Q_{wi} = min [\sum ^{M}_{m=1} G_{mi}, \sum ^{N}_{n=1} U_{ni}]  $$
    
$$ q_{mi} = Q_{wi} \times \frac{G_{mi}}{\sum ^{M}_{m=1} G_{mi}}  $$    
        
$$ q_{ni} = Q_{wi} \times \frac{U_{ni}}{\sum ^{N}_{n=1} U_{ni}}  $$


使用  dummy variable來處理

$$ U_{ni}=min[A_{ni}, C_{ni}]$$

$$ Q_{wi} = min [\sum ^{M}_{m=1} G_{mi}, \sum ^{N}_{n=1} U_{ni}]  $$

$$ y_{ni}( 1 - y_{ni} ) = 0 $$

$$ y_{wi}( 1 - y_{wi} ) = 0 $$

將 $ U_{ni} = min[A_{ni}, C_{ni}] $ 轉化為:

$$ C_{ni} \geq  U_{ni} $$

$$ A_{ni} \geq  U_{ni} $$

$$ U_{ni} \geq A_{ni} - k_{ni} (1-y_{ni}) $$

$$ U_{ni} \geq C_{ni} - k_{ni} \times y_{ni} $$

將 $ Q_{wi} = min [\sum ^{M}_{m=1} G_{mi}, \sum ^{N}_{n=1} U_{ni}] $ 轉化為:

$$ \sum ^{M}_{m=1} G_{mi} \geq  Q_{wi} $$

$$ \sum ^{N}_{n=1} U_{ni} \geq  Q_{wi} $$

$$ Q_{wi} \geq \sum ^{M}_{m=1} G_{mi} - k_{ni} (1-y_{wi}) $$

$$ Q_{wi} \geq \sum ^{N}_{n=1} U_{ni} - k_{ni} y_{wi} $$

五項轉供式可以寫為:

$$ G_{mi} = R_{m} \times P_{mi} $$
    
$$ C_{ni} \geq  U_{ni} $$

$$ A_{ni} \geq  U_{ni} $$

$$ U_{ni} \geq A_{ni} - k_{ni} (1-y_{ni}) $$

$$ U_{ni} \geq C_{ni} - k_{ni} \times y_{ni} $$
    
$$ \sum ^{M}_{m=1} G_{mi} \geq  Q_{wi} $$

$$ \sum ^{N}_{n=1} U_{ni} \geq  Q_{wi} $$

$$ Q_{wi} \geq \sum ^{M}_{m=1} G_{mi} - k_{ni} (1-y_{wi}) $$

$$ Q_{wi} \geq \sum ^{N}_{n=1} U_{ni} - k_{ni} y_{wi} $$

$$ y_{ni}( 1 - y_{ni} ) = 0 $$

$$ y_{wi}( 1 - y_{wi} ) = 0 $$

$$ q_{mi} = Q_{wi} \times \frac{G_{mi}}{\sum ^{M}_{m=1} G_{mi}}  $$    
        
$$ q_{ni} = Q_{wi} \times \frac{U_{ni}}{\sum ^{N}_{n=1} U_{ni}}  $$

    
太棒了，變成標準型惹，衰敗且收斂。

# demo 2
#計算  (2+x1)/(1+x2) - 3*x1+4*x3 的最小值  x1,x2,x3的範圍都在0.1到0.9 之間
def fun(args):
    a,b,c,d=args
    v=lambda x: (a+x[0])/(b+x[1]) -c*x[0]+d*x[2]
    return v
def con(args):
    # 約束條件 分爲eq 和ineq
    #eq表示 函數結果等於0 ； ineq 表示 表達式大於等於0  
    x1min, x1max, x2min, x2max,x3min,x3max = args
    cons = ({'type': 'ineq', 'fun': lambda x: x[0] - x1min},\
              {'type': 'ineq', 'fun': lambda x: -x[0] + x1max},\
             {'type': 'ineq', 'fun': lambda x: x[1] - x2min},\
                {'type': 'ineq', 'fun': lambda x: -x[1] + x2max},\
            {'type': 'ineq', 'fun': lambda x: x[2] - x3min},\
             {'type': 'ineq', 'fun': lambda x: -x[2] + x3max})
    return cons
 
if __name__ == "__main__":
    #定義常量值
    args = (2,1,3,4)  #a,b,c,d
    #設置參數範圍/約束條件
    args1 = (0.1,0.9,0.1, 0.9,0.1,0.9)  #x1min, x1max, x2min, x2max
    cons = con(args1)
    #設置初始猜測值  
    x0 = np.asarray((0.5,0.5,0.5))
    
    res = opt.minimize(fun(args), x0, method='SLSQP',constraints=cons)
    print(res.fun)
    print(res.success)
    print(res.x)
